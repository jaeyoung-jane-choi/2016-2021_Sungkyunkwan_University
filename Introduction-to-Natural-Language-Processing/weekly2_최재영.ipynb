```python
#Frequency Distributions 
fdist = FreqDist(text5)
print(fdist) #unique 6066, total 45010 : same as above 
```

```python
fdist.most_common(10) #dictionary result : top 10 frequent words in fdist 
```

```python
fdist['lol'] #dictionary: key-value link 
```

```python
fdist.plot(20, cumulative = True) #top 20 accounts to 12000 words 
```

```python
#fine-grained Selection of Words 

sorted(w for w in set(text5) if len(w)>5 and fdist[w]>10)
```

```python
#Collocations : sequence of words that occur together 
text5.collocations()
```

```python
#Counting word lengths-distribution 
fdist2 = FreqDist(len(w) for w in text5)
fdist2.most_common() #3 letter word is most common 
```

```python
fdist2.freq(3) #24% is 3 letter word
```

```python
#Counting Words by ConditionalFreqDist()

genre_word = [(genre,word)
              for genre in ['romance', 'adventure'] 
              for word in nltk.corpus.brown.words(categories = genre)]
genre_word[:10]
```

```python
genre_word[-10:]
```

```python
cfd = nltk.ConditionalFreqDist(genre_word)

cfd['adventure'].most_common(10) #dictionary with value of tuple of list / key is category 
```
