```
import nltk
nltk.download('movie_reviews')
nltk.download('stopwords')
nltk.download('names')
from nltk.corpus import movie_reviews
import random 
```

```
#category: each review as positive or negative 
documents = [(list(movie_reviews.words(fileid)),category) for category in movie_reviews.categories() for fileid in movie_reviews.fileids(category)]
documents[0][1]
# documents : [wordslist, category] 
```

```
all_words = nltk.FreqDist(w.lower() for w in movie_reviews.words())
word_features = list(all_words)[:2000] #list all the words in movie_reviews and use 2000 words as word features 
word_features[:10]
#len(word_features)
```

```
def document_features(document):
  document_words =set(document) #unique words of the document 
  features = {}
  for word in word_features:
    features[word] = (word in document_words)
  return features


  #returns a dictionary of whether the word in a document has words in the word_feature list 
```

Lets not use True or false , just use total numbers of occurance 


```
update_all_words['plot']
```

```
sorted_words = sorted(update_all_words.items(), key = lambda item: item[1], reverse=True)
top3000 = sorted_words[:3000]

top3000_dict = {}
for item in top3000:
  top3000_dict[item[0]] = item[1]


len(top3000_dict)
```

```


def second_document_features(document):
  document_words = document
  fdist = nltk.FreqDist(document)
  #print(fdist.most_common(10))
  features ={}
  for key in top3000_dict.keys():
    if key in fdist.keys():
      features[key] = fdist[key]
    else:
      features[key] = 0
    

  return features

print(second_document_features(movie_reviews.words('pos/cv957_8737.txt'))) 
```

```
second_featuresets = [(second_document_features(d), c) for (d,c) in documents]
```

```

second_train, second_test = second_featuresets[100:], second_featuresets[:100]

second_classifier = nltk.NaiveBayesClassifier.train(second_train) #train with NBClassifier 
new_second_classifier_accuracy =nltk.classify.accuracy(second_classifier,second_test)

print(new_second_classifier_accuracy) #test with the accuracy 
```

Lets upgrade the accuracy by using laplace smoothing (similar; not added total unique) 

```


def new_second_document_features(document):
  document_words = document
  fdist = nltk.FreqDist(document)
  #print(fdist.most_common(10))
  features ={}
  for key in top3000_dict.keys():
    if key in fdist.keys():
      features[key] = fdist[key]+ 1 
    else:
      features[key] = 1
    #add one to make non-zeros 

  return features

second_featuresets = [(new_second_document_features(d), c) for (d,c) in documents]

second_train, second_test = second_featuresets[100:], second_featuresets[:100]

second_classifier = nltk.NaiveBayesClassifier.train(second_train) #train with NBClassifier 
laplace_second_classifier_accuracy =nltk.classify.accuracy(second_classifier,second_test)

print(laplace_second_classifier_accuracy) #test with the accuracy 
```
